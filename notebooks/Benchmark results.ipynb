```python
import pickle

max_sites = 200
step = 50
num_samples = 3

with open(f"../results_{max_sites}_{step}_{num_samples}.pkl", "rb") as file:
    results = pickle.load(file)
```

```python
x_axis = []
dw_vals = []
sa_vals = []
sb_vals = []

for k in ['50', '100']:
    for i, val in enumerate(results[k]['sa']['energies']):
        sa_vals.append(val)
        x_axis.append(f"{k}.{i}")
    for i, val in enumerate(results[k]['dwave']['energies']):
        if val:
            dw_vals.append(val)
        else:
            dw_vals.append(0)
    for i, val in enumerate(results[k]['sb']['energies']):
        sb_vals.append(val)
```

```python
import numpy as np
from ising_benchmark.utils import comparison_plot

arrays = np.array([sa_vals, sb_vals, dw_vals])

comparison_plot(arrays, ["Simulated annealing", "Simulated bifurcation","DWave hybrid solver"], x_ticks=x_axis)
```
